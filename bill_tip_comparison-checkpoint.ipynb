```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('seaborn')
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import linear_model
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df = df.loc[:, 'total_bill':'tip']
```

```python
df
```

```python
plt.figure(figsize = (15, 10))
sns.scatterplot(data = df['tip'], s = 100)
sns.lineplot(df.index, df['tip'].mean(), linewidth = 5, color = 'r')
plt.show()
```

```python
plt.figure(figsize = (15, 10))
sns.scatterplot(df['total_bill'], df['tip'], s = 100)
sns.lineplot(df['total_bill'], df['tip'].mean(), linewidth = 5, color = 'r')
plt.show()
```
