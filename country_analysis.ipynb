```python
data = pd.read_csv(r'hdi_master.csv')
```

```python
data
```

```python
missing = data['HDI for year'][data['year'] == 2010].isnull()
hdi = np.array(data['HDI for year'][data['year'] == 2010][~missing])
hdi
```

```python
hdi.shape
```

```python
plt.figure(figsize = (15, 5))

plt.hist(hdi, facecolor = 'orange', alpha = 0.5)

plt.axvline(hdi.mean(), label = 'mean', linestyle = 'solid', linewidth = 2)
plt.axvline(hdi.mean() + hdi.std(), label = '+ STD', linestyle = 'solid', linewidth = 2)
plt.axvline(hdi.mean() - hdi.std(), label = '- STD', linestyle = 'solid', linewidth = 2)

plt.ylabel('HDI Count')
plt.xlabel('HDI')
plt.title('HDI Count for 2010')
plt.legend(loc = 'upper left')


plt.show()
```
